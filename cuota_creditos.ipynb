```python
tasa = 0.04
n_cuotas = 60
nominal = 1000000
```

```python
valor_cuota = nominal * ((tasa/12) * (1 + (tasa/12))**n_cuotas)/((1 + (tasa/12))**n_cuotas - 1)
valor_cuota
```

```python
amort = [0 for i in range(n_cuotas)]
sum_amort = [0 for i in range(n_cuotas+1)]
intereses = [0 for i in range(n_cuotas)]
```

```python
for i in range(n_cuotas):
    intereses[i] = (nominal - sum_amort[i]) * (tasa/12)
    amort[i] = valor_cuota - intereses[i]
    sum_amort[i+1] = sum_amort[i] + amort[i]
```
